```python
tabelas = {}
# importar tabela das pastas
def importar_tabelas(pasta):
    for item in os.listdir(pasta):
        item_path = os.path.join(pasta, item)
        if os.path.isdir(item_path):
            importar_tabelas(item_path)  # Chamada recursiva para subpastas
        elif item.endswith('.csv'):  # Verifica se o item é um arquivo CSV
            df = pd.read_csv(item_path)  # Importa o arquivo CSV como DataFrame
            # Faça o que precisar com o DataFrame (análise, processamento, etc.)
            print(f'Conteúdo da tabela {item}:')
            print(df)
            print('\n')
```
